## Linear Probing

```python
# imports
import re
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
```

### Multi-Channel S2 data

#### Linear Probing on trained I-JEPA

```python
# Load the log file
log_file_path = "jepa_probing.txt"
with open(log_file_path, "r", encoding="utf-8") as f:
    lines = f.readlines()

# Lists to hold the metrics
epochs = []
train_losses = []
val_losses = []
val_accuracies = []

# Regex pattern to extract values
pattern = r"Epoch \[(\d+)/\d+\] Train Loss: ([\d\.]+) Val Loss: ([\d\.]+) Val Acc: ([\d\.]+)%"

# Extract data
for line in lines:
    match = re.search(pattern, line)
    if match:
        epoch = int(match.group(1))
        train_loss = float(match.group(2))
        val_loss = float(match.group(3))
        val_acc = float(match.group(4))
        
        epochs.append(epoch)
        train_losses.append(train_loss)
        val_losses.append(val_loss)
        val_accuracies.append(val_acc)

```

```python
# Plot the metrics
plt.figure(figsize=(10, 6))
plt.plot(epochs, train_losses, label="Train Loss", marker='o')
plt.plot(epochs, val_losses, label="Validation Loss", marker='o')

plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Loss Over Epochs (I-JEPA Linear Probing with Pretrained Model)")
plt.gca().xaxis.set_major_locator(MaxNLocator(integer=True))
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Plot the metrics
plt.figure(figsize=(10, 6))
plt.plot(epochs, val_accuracies, label="Validation Accuracy (%)", marker='o')

plt.xlabel("Epoch")
plt.ylabel("Accuracy (%)")
plt.title("Validation Accuracy Progression (I-JEPA Linear Probing with Pretrained Model)")
plt.gca().xaxis.set_major_locator(MaxNLocator(integer=True))
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

## SSL Training

### I-JEPA trained on multi-channel S2 data

```python
log_path = 'i-jepa.log'

# Lists to store parsed losses
train_losses = []
val_losses = []
val_steps = []

# Read and parse the log
with open(log_path, 'r', encoding='utf-8') as file:
    step = 0
    for line in file:
        # Match training loss
        match_train = re.search(r'loss: ([0-9.]+)', line)
        if match_train and 'Validation' not in line:
            train_losses.append(float(match_train.group(1)))
            step += 1
        
        # Match validation loss
        match_val = re.search(r'Validation loss: ([0-9.]+)', line)
        if match_val:
            val_losses.append(float(match_val.group(1)))
            val_steps.append(step)  # Save step index for x-axis

# Create x-axis values
train_steps = list(range(1, len(train_losses) + 1))

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(train_steps, train_losses, label='Training Loss', alpha=0.7)
plt.plot(val_steps, val_losses, label='Validation Loss (every 120 steps)', marker='o', linestyle='--')
plt.xlabel('Step')
plt.ylabel('Loss')
plt.title('Training and Validation Loss (I-JEPA SSL Training on Sentinel-S2 Multi-Channel Data)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
